```python
digit_numbers = 3
digits = [9 for x in range(1,digit_numbers+1)]
place_numbers = '9'*digit_numbers
place_numbers = int(place_numbers)
best_palindrome = 0 

numerics = []
for i in digits:
    for n in range(1, place_numbers+1):
        for m in range(1, place_numbers+1):
            out = n*m 
            if len(str(out))%2==0:
                revers = (str(out))[::-1]
                if str(out)== revers:
                    numerics.append(out)
                    if out > best_palindrome: best_palindrome = out
numerics
```

```python
digit_numbers = 7

place_numbers = int('9'*digit_numbers)

best_palindrome = 0 

for i in range(place_numbers, 9, -1):
    for j in range(i, 9, -1):
        prodkt = i * j 
        if prodkt <= best_palindrome:
            break
        if str(prodkt) == str(prodkt)[::-1]:
            best_palindrome = prodkt
```
