```python
def gcd(a, b):
    # Compute the greatest common divisor (GCD) using Euclid's algorithm
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    # Compute the least common multiple (LCM) using GCD
    return a * b // gcd(a, b)

def find_smallest_multiple(n):
    # Start with the first number and compute LCM iteratively for all numbers in range
    result = 1
    for num in range(1, n + 1):
        print(result,num)
        result = lcm(result, num)
    return result

number_range = 20  # Dynamic variable
smallest_multiple = find_smallest_multiple(number_range)
print(smallest_multiple)
```
